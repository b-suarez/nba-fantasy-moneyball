```python
# Import libraries
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

## Data Reading

First of all we will read the biwinger API to get the values of the players in the platform

```python
# We get the response from the API
response = requests.get("https://cf.biwenger.com/api/v2/competitions/nba/data").json()
```

```python
# Get the data portion
data = response["data"]

# Get the dict from the players
df_dict = pd.DataFrame.from_dict(data["players"])
```

When reading the data from this kind of dictionaries we need to transpose it, so the columns look how we want them to.

```python
# We need to transpose de df
players_df = df_dict.transpose()

# We are removing the index column (we already have ID)
players_df.reset_index(drop=True,inplace=True)

# Show the head of the transformed head
players_df.head()
```

## Generating the new variables

In order to make a proper evaluation of the players, we need to generate some columns.

- <b>Points per game:</b> We will use the the average of the points scored per game for each player
- <b>points/price:</b> This is the metric we will use to check how the price of a player correspond to how much it costs

Let's do this!

```python
# Create the total games played column
players_df["gamesPlayed"] = players_df["playedHome"] + players_df["playedAway"]

# Total points (we need this column because the points column represents the points per
# week and we need the points per game)
players_df["totalPoints"] = players_df["pointsHome"] + players_df["pointsAway"]

# Create the pointsPerGame column
def calculate_points_per_game(row):
    if row["gamesPlayed"] > 0:
        return row["totalPoints"]/row["gamesPlayed"]
    else:
        return 0

# We apply the previous function to each row of the df
players_df["pointsPerGame"] = players_df.apply(calculate_points_per_game, axis=1)
```

```python
# We generate the points per coin ratio column
players_df["pointsPerCoin"] = players_df["pointsPerGame"] / players_df["price"] * 1000

x = players_df[['pointsPerCoin']].values.astype(float)
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)

players_df["pointsPerCoin"] = x_scaled
```

```python
players_df.head()
```

Ler's plot the points per game / price to get an idea of how this new variable works

```python
fig, ax = plt.subplots(figsize=(20, 20))
ax.scatter(players_df["pointsPerGame"], players_df["price"])
plt.xlabel("Points Per Game")
plt.ylabel("Price")

for i, txt in enumerate(players_df["name"]):
    ax.annotate(txt,(players_df["pointsPerGame"][i], players_df["price"][i]))
```

Let's create a discretized column of the points per game column for better clasification.

```python
# Let's define the ranges and bins
bins = pd.IntervalIndex.from_tuples([(-10, 0), (0, 5), (5, 10), (10, 15), (15, 20), (20, 25), (25, 30), (30, 35), (35, 40), (40, 45)])
players_df["range_ppg"] = pd.cut(players_df["pointsPerGame"], bins)

# We are removing the index column and using the rank in pointPerCoin
players_df = players_df.sort_values(by=['pointsPerCoin'], ascending=False)
players_df.reset_index(drop=True,inplace=True)
```

## Let's check for the best players per value

In my case, I'm interested in players with an average of points > 10 due to it being the average I am looking for in a bench player.

```python
useful_players_df = players_df.loc[players_df["pointsPerGame"]>10]
# We are removing the index column and using the rank in pointPerCoin

players_df = players_df.sort_values(by=['price'], ascending=False)
players_df.reset_index(drop=True,inplace=True)
```

## Evaluate Players

```python
def print_efficiency_rank(player_name):
    """ Prints the rank of a player in the value ranking
    and shows how good its value is against the mean price
    of players in its PPG range """
    player_row = players_df.loc[players_df["name"] == player_name]
    player_eff_range = player_row["range_ppg"].values
    lol = players_df.loc[players_df["range_ppg"] == player_eff_range[0]]["pointsPerCoin"].mean()
    print('{} is ranked {} in overal value and has a ratio {} above/below the mean of the players in the range {} PPG'.format(player_name,
                               player_row.index.tolist(),
                               player_row["pointsPerCoin"].values - players_df.loc[players_df["range_ppg"] == player_eff_range[0]]["pointsPerCoin"].mean(),
                               player_eff_range[0]                                                                                       
                              ))
```
